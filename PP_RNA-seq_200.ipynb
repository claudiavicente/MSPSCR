```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import umap
```

```python
path = "/Users/claudiavicentecomorera/Desktop/TFG/Data/RNA_seq/"

hippocampus = pd.read_csv(os.path.join(path, "gene_tpm_brain_hippocampus.gct"), index_col = 2, skiprows = 2, sep = '\t')
lung = pd.read_csv(os.path.join(path, "gene_tpm_lung.gct"), index_col = 2, skiprows = 2, sep = '\t')

hippocampus = hippocampus.iloc[:, 2:]
lung = lung.iloc[:, 2:]
```

```python
log_hippocampus = np.log2(hippocampus + 1)
log_lung = np.log2(lung + 1)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 10))

sns.kdeplot(data = log_hippocampus, ax = axes[0], common_norm = False, legend = False, palette = "gnuplot2", linewidth = .2, alpha = .6)
axes[0].set_title('Hippocampus')
sns.kdeplot(data = log_lung, ax = axes[1], common_norm = False, legend = False, palette = "gnuplot2", linewidth = .2, alpha = .6)
axes[1].set_title('Lung')

plt.suptitle('Density Plots of samples gene expression')
plt.tight_layout()
```

```python
df = pd.concat([log_hippocampus, log_lung], axis = 1)
df
```

```python
t_idx = np.cumsum([len(log_hippocampus.columns), len(log_lung.columns)])
t_labels = ['Hippocampus', 'Lung']
colors = ['r', 'b']

reducer = umap.UMAP()
embedding = reducer.fit_transform(df.T)

plt.figure(figsize = (10, 8))
for i, (start, end) in enumerate(zip([0] + list(t_idx), t_idx)):
    plt.scatter(embedding[start:end, 0], embedding[start:end, 1], color = colors[i], label = t_labels[i], alpha = .3)
plt.title('UMAP projection of gene expression from different tissues')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.legend()
plt.show()
```

```python
s_loghippocampus = log_hippocampus.sample(n = 100, axis = 1, random_state = 42)
s_loglung = log_lung.sample(n = 100, axis = 1, random_state = 42)

s_df = pd.concat([s_loghippocampus, s_loglung], axis = 1)

embedding = reducer.fit_transform(s_df.T)
colors = ['r'] * 100  + ['b'] * 100

plt.figure(figsize = (10, 8))
plt.scatter(embedding[:, 0], embedding[:, 1], c = colors, alpha = .3)
plt.title('UMAP projection of gene expression of 200 random samples from different tissues')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

```python
df = pd.concat([hippocampus, lung], axis = 1)
df[s_df.columns].to_csv('/Users/claudiavicentecomorera/Desktop/TFG/Data/RNA_seq/bulk_data200.csv')
df[s_df.columns]
```
