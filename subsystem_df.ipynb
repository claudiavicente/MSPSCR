```python
import cobra
import pandas as pd
```

```python
recon3D = cobra.io.load_matlab_model('/Users/claudiavicentecomorera/Desktop/TFG/Data/Recon3D_301/Recon3DModel_301.mat')
recon3D
```

```python
IDs = []
subsystems = []

for rx in recon3D.reactions:
    IDs.append(rx.id)
    subsystems.append(recon3D.get_associated_groups(rx))

subsytems_df = pd.DataFrame({'ID': IDs, 'Subsystem': subsystems})
subsytems_df
```
