# FLUXES SIMULATION

## SETTING FILE PATHS

```python
f_path = "/Users/claudiavicentecomorera/Desktop/TFG/Models/PBMC_400.2Models/"
file_paths = glob.glob(f_path + "*.mat")
```

## pFBA

```python
models = []
solutions = []
times = []

for file_path in file_paths:
    model = cobra.io.load_matlab_model(file_path)
    model.id = os.path.splitext(os.path.basename(file_path))[0].replace("_", "-")
    
    t0 = time.time()  
    model.objective = 'biomass_maintenance'
    solution = cobra.flux_analysis.pfba(model) 
    t = time.time()  
    pfba_time = t - t0
    
    print(f"The simulation of {model.id} fluxes takes: {pfba_time} seconds")
    
    models.append(model)
    solutions.append(solution)
    times.append(pfba_time)
```

```python
r_models = pd.DataFrame(index = [model.id for model in models])

for i, solution in enumerate(solutions, start = 0):
    ov_m = solution.objective_value
    percent = (ov_m/ov) * 100

    r_models.loc[r_models.index[i], 'biomass_maintenance'] = solutions[i].fluxes["biomass_maintenance"]
    r_models.loc[r_models.index[i], 'objective_value'] = ov_m
    r_models.loc[r_models.index[i], 'percent_ov'] = percent
    r_models.loc[r_models.index[i], 'feasible_models'] = "TRUE" if percent > 25 else "FALSE"


print(f"The 25% of Recon 3D objective value is {25/100*ov}")
print(f"There are {r_models['feasible_models'].eq('FALSE').sum()} models that aren't feasible because of their low objective value.\n")
r_models.head()
```

```python
fr_models = r_models[r_models['feasible_models'] == 'TRUE']
fr_models
```

# RESULTS VISUALIZATION

## CREATION OF THE FLUXOMES DATAFRAME

```python
fluxes_df = pd.DataFrame(index = [model.id for model in models])

for i, solution in enumerate(solutions, start = 0):
    
    fluxes = solution.fluxes
    
    for reaction, flux in fluxes.items():
        if reaction not in fluxes_df.columns:
            fluxes_df[reaction] = 0.0
        
        fluxes_df.loc[fluxes_df.index[i], reaction] = flux
        fluxes_df = fluxes_df.copy()

fluxes_df.head()
```

## UMAP PROJECTION

```python
c_labs = pd.read_csv('/Users/claudiavicentecomorera/Desktop/TFG/Data/PBMC/clabs_pbmc400_2.csv')
d_clabs = {}

for index, row in c_labs.iterrows():
    cell = row['Cell_Name']
    type = row['Cluster_Label']

    if type in d_clabs:
        d_clabs[type].append(cell)
    else:
        d_clabs[type] = [cell]
```

```python
l = []
l_labs = []
for key, value in d_clabs.items():
    if key == 'CD14 Monocytes':
        l.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#ff7f0e', markersize=10))
    elif key == 'B':
        l.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#279e68', markersize=10))
    l_labs.append(key)
```

```python
colors = []

for cell in fluxes_df.index:
    for type in d_clabs:
    
        for cells in d_clabs[type]:
            if cell in cells:
                if type == 'CD14 Monocytes':
                    colors.append('#ff7f0e')
                elif type == 'B':
                    colors.append('#279e68')
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(fluxes_df)

plt.scatter(embedding[:, 0], embedding[:, 1], c = colors) 
plt.legend(l, l_labs)
plt.title('UMAP of 400 PBMC models fluxes distributions')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

```python
d_ov = dict(zip(r_models.index, r_models['percent_ov']))

plt.scatter(embedding[:, 0], embedding[:, 1], c = [d_ov.get(index, -1) for index in fluxes_df.index], cmap = "viridis")
plt.colorbar(label = 'Percentage of Recon 3D objective value')  
plt.title('UMAP of 400 PBMC models fluxes distributions')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

# RESULTS VISUALIZATION WITH TRANSFORMED FLUXES

```python
def flux_transformer(v):
    
    '''
    This function transforms fluxes into log2 scale with three additions:
    - zero values are kept.
    - values between 0 and 2 are adjusted to fit below one.
    - negative fluxes are considered. 
    '''

    def adjusted(v):
        new = (np.log2(v + 0.125) + 3) / 4
        return np.min([new, 1])

    if v == 0:
        vprime = 0
    elif v > 0:
        if v < 2:
            vprime = adjusted(v)
        else:
            vprime = np.log2(v)
    elif v < 0:
        v = np.abs(v)
        if v < 2:
            vprime = -adjusted(v)
        else:
            vprime = -np.log2(v)
    else:
        raise ValueError('found value not considered')

    return vprime
```

```python
t_fluxes_df = fluxes_df.applymap(flux_transformer)
t_fluxes_df
```

```python
reducer = umap.UMAP(n_neighbors = 10, local_connectivity = 1)
embedding = reducer.fit_transform(t_fluxes_df)

plt.scatter(embedding[:, 0], embedding[:, 1], c = colors) 
plt.legend(l, l_labs)

plt.title('UMAP of 400 PBMC models transformed fluxes distributions')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

```python
plt.scatter(embedding[:, 0], embedding[:, 1], c = [d_ov.get(index, -1) for index in fluxes_df.index], cmap = "viridis")
plt.colorbar(label = 'Biomass maintenance flux')  
plt.title('UMAP of 400 PBMC models transformed fluxes distributions')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```
